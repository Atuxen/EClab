```python
file_dict = {
    "Zn Steel Molal 1": "./ZnSO4 characterization week2/Cycling/ZnSO4-Molal-1-Coincell-50mikroL-Zn-Steel_02_GCPL_C06.mpr",
    "Zn Steel Molal 2": "./ZnSO4 characterization week2/Cycling/ZnSO4-Molal-2-Coincell-50mikroL-Zn-Steel_02_GCPL_C01.mpr",
}


def CEplot(dict):
    output_notebook()
    CE = figure(
        title="Coloumbic Effciency",
        x_axis_label="Cycle number",
        y_axis_label="Coloumbic Effciency (%)",
        width=800,
        height=400,
    )

    ## Constrain axes
    # CE.x_range.start = -0.5  # Set x-axis range from 0 to 3V
    # CE.x_range.end = 4
    # CE.y_range.start = -20  # Set y-axis range from 0 to 20 mA/cm^2
    # CE.y_range.end = 20

    color = 0
    for key, value in dict.items():

        print(value)

        df2 = ecf.to_df(value)
        # name = path.split('/')[-1]

        # Take the absolute values of `Q charge/discharge`
        df2["Q charge/discharge"] = df2["Q charge/discharge"].abs()

        # Find the maximum charge and discharge for each half cycle
        max_charge = (
            df2[df2["ox/red"] == 1].groupby("half cycle")["Q charge/discharge"].max()
        )
        max_discharge = (
            df2[df2["ox/red"] == 0].groupby("half cycle")["Q charge/discharge"].max()
        )

        # Align the indices of charge and discharge cycles for a full cycle comparison
        # Assuming that 'half cycle' pairs consecutive charge and discharge cycles (e.g., 0 and 1, 2 and 3)
        charge_cycles = max_charge.iloc[
            :-1
        ]  # Exclude the last half cycle if it's incomplete
        discharge_cycles = max_discharge.iloc[
            1:
        ]  # Exclude the first half cycle if it lacks a discharge

        # Reset indices for alignment
        charge_cycles = charge_cycles.reset_index(drop=True)
        discharge_cycles = discharge_cycles.reset_index(drop=True)

        # Calculate Coulombic Efficiency
        coulombic_efficiency = (discharge_cycles / charge_cycles) * 100

        # Combine into a single DataFrame
        full_cycles = pd.DataFrame(
            {
                "Charge Capacity": charge_cycles,
                "Discharge Capacity": discharge_cycles,
                "Coulombic Efficiency (%)": coulombic_efficiency,
            }
        )

        cycle = charge_cycles  # df['half cycle']
        coleff = coulombic_efficiency

        CE.line(
            cycle,
            coleff,
            legend_label=key,
            line_color=small_palettes["Viridis"][4][color],
            line_width=2,
        )

        color = color + 1
        print(color)

    CE.legend.location = "top_left"
    CE.legend.title = "ZnSO4 Concentrations"
    show(CE)


CEplot(file_dict)
```

```python
import pandas as pd
from bokeh.plotting import figure, output_notebook, show
from bokeh.palettes import Viridis256

# Define file paths
file_dict = {
    "Zn Steel Molal 1": "./ZnSO4 characterization week2/Cycling/ZnSO4-Molal-1-Coincell-50mikroL-Zn-Steel_02_GCPL_C06.mpr",
    "Zn Steel Molal 2": "./ZnSO4 characterization week2/Cycling/ZnSO4-Molal-2-Coincell-50mikroL-Zn-Steel_02_GCPL_C01.mpr",
}

# Example `ecf.to_df` must be imported/defined
# Import your library or function to handle .mpr files here
# Example:
# import ecf

# Bokeh color palette
small_palettes = {"Viridis": [Viridis256]}


def CE(file_dict):
    output_notebook()
    CE = figure(
        title="Coulombic Efficiency",
        x_axis_label="Cycle number",
        y_axis_label="Coulombic Efficiency (%)",
        width=800,
        height=400,
    )

    color = 0
    for key, value in file_dict.items():
        print(value)

        # Convert .mpr file to DataFrame
        df2 = ecf.to_df(value)

        # Take absolute values of Q charge/discharge
        df2["Q charge/discharge"] = df2["Q charge/discharge"].abs()

        # Find maximum charge and discharge per half cycle
        max_charge = (
            df2[df2["ox/red"] == 1].groupby("half cycle")["Q charge/discharge"].max()
        )
        max_discharge = (
            df2[df2["ox/red"] == 0].groupby("half cycle")["Q charge/discharge"].max()
        )

        # Align indices for full-cycle comparison
        charge_cycles = max_charge.iloc[:-1].reset_index(drop=True)
        discharge_cycles = max_discharge.iloc[1:].reset_index(drop=True)

        # Calculate Coulombic Efficiency
        coulombic_efficiency = (discharge_cycles / charge_cycles) * 100

        # Extract cycle numbers
        cycle = charge_cycles.index
        coleff = coulombic_efficiency

        # Add line to Bokeh plot
        CE.line(
            cycle,
            coleff,
            legend_label=key,
            line_color=small_palettes["Viridis"][0][color],
            line_width=2,
        )

        color += 1

    CE.legend.location = "top_left"
    CE.legend.title = "ZnSO4 Concentrations"
    show(CE)


# Call the function
CE(file_dict)
```

# Potentiostatic Electrochemical Impedance Spectroscopy (PEIS)

```python
from bokeh.io import output_notebook, show
from bokeh.layouts import row
from bokeh.plotting import figure

# import pandas as pd

df = ecf.to_df(
    "./ZnTFSI2 characterization/ZnTFSI2-G4-1-5-Coincell-60mikroL-Zn-Steel_01_PEIS_C01.mpr"
)
print(df.columns)

potential = df["<Ewe>"]
time = df["time"]

zReal = df["Re(Z)"]
zImagine = df["-Im(Z)"]


output_notebook()

nyquist_plot = figure(
    title="Nyquist Plot",
    x_axis_label="Z' (Ω)",
    y_axis_label="-Z'' (Ω)",
    width=400,
    height=300,
)
nyquist_plot.line(
    zReal, zImagine, legend_label="Nyquist Curve", line_color="blue", line_width=2
)

e_vs_i_plot = figure(
    title="Ewe vs. Time Plot",
    x_axis_label="Time (h)",
    y_axis_label="Ewe (V)",
    width=400,
    height=300,
)
e_vs_i_plot.line(
    time,
    potential,
    legend_label="E vs. Time",
    line_color="red",
    line_width=2,
    line_dash="dashed",
)

layout = row(nyquist_plot, e_vs_i_plot)
show(layout)
```

# Coloumbic Effciency 
Coulombic Efficiency(%)= 
Charge Capacity/Discharge Capacity ×100

```python
import pandas as pd

# Make a copy of the DataFrame
df = ecf.to_df(
    "./ZnTFSI2 characterization/ZnTFSI2-G4-1-5-Coincell-60mikroL-Zn-Steel_02_GCPL_C01.mpr"
)

# Take the absolute values of `Q charge/discharge`
df["Q charge/discharge abs"] = df["Q charge/discharge"].abs()

# Find the maximum charge/discharge for each half cycle So that we have the peak charges
full_charge_cycles = df.groupby("half cycle")["Q charge/discharge abs"].max()


# Initialize a DataFrame to store results
results = pd.DataFrame(
    {
        "FullchargeCyles": full_charge_cycles.values,
    }
)

# Calculate Coulombic Efficiency
# The CE for cycle `i` depends on the charge/discharge capacity of cycle `i-1`
results["Cyclecharge"] = results["FullchargeCyles"].shift(1) + results[
    "FullchargeCyles"
].shift(0)
results["ColEff"] = (
    results["Cyclecharge"].shift(1) / results["Cyclecharge"].shift(0) * 100
)

# Handle the first cycle (no CE calculation possible)
results.loc[0, "ColEff"] = None  # Optional, could also be set to 0


output_notebook()
cycle = list(range(0, 1002))  # df['half cycle']
coleff = results["ColEff"]

GCPL = figure(
    title="Galvanostatic Cycling with Potential Limitation",
    x_axis_label="Cycle (h)",
    y_axis_label="Ewe (V vs Zn)",
    width=800,
    height=300,
)
GCPL.line(cycle, coleff, legend_label="Ewe vs. Time", line_color="blue", line_width=2)

show(GCPL)
```
